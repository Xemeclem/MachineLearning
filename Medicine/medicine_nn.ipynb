```python
import torch
import rdkit
import numpy as np
import pandas as pd
import deepchem as dc
import matplotlib.pyplot as plt
from torch import nn, optim
from torch.utils.data import Dataset, DataLoader
from rdkit import Chem
from torch.nn.parameter import Parameter
import tqdm, sys, os, gc, re, argparse, warnings, math, random
warnings.filterwarnings("ignore",category=DeprecationWarning)
```

```python
# 数据预处理，excel文件 ----> DataFrame类
train = pd.read_excel('medicine_dataset/dataset-new/traindata-new.xlsx')
test = pd.read_excel('medicine_dataset/dataset-new/testdata-new.xlsx')

# 选择有意义的特征
cols_physics = ['Molecular Weight', 'Exact Mass', 'XLogP3', 'Heavy Atom Count',
             'Ring Count', 'Hydrogen Bond Acceptor Count', 'Hydrogen Bond Donor Count',
             'Rotatable Bond Count', 'Topological Polar Surface Area']
cols_chemistry = ['Smiles']
cols_text = ['Target', 'E3 ligase']
```

```python
## 特征工程，使用 DeepChem 函数库对分子进行特征化
mol = [Chem.MolFromSmiles(i) for i in np.array(train['Smiles'])]

# CircularFingerprint特征化方法，将分子的 Smiles 转化成机器学习需要的向量形式
# mol对象--->Numpy数组 (shape：分子的个数 x 1024, dtype: Double)
cir_featurizer = dc.feat.CircularFingerprint(size=1024, radius=2)
cir_features = cir_featurizer.featurize(mol)

# RDKitDescriptors特征化方法，提取分子的一些基本的物理和化学特性：如分子量，极性表面积，氢键供体和受体的数量等
# mol对象--->Numpy数组 (shape：分子的个数 x 210, dtype: Double)
rdkit_featurizer = dc.feat.RDKitDescriptors()
rdkit_features = rdkit_featurizer(mol)

data_train = np.concatenate((cir_features, rdkit_features, np.array(train[cols_physics])), axis=1) # shape: 分子的个数 x (1024 + 210 + 9), dtype: Double
data_label = np.array(train['Label']) # len: 分子个数, dtype: long int
```

```python
# Numpy数组--->Tensor: float, int
class Mydata(Dataset):
    def __init__(self, x, y):
        self.inputs = torch.tensor(x, dtype = torch.float)
        self.labels = torch.tensor(y, dtype = torch.long)
          
    def __getitem__(self, item):
        return self.inputs[item,:], self.labels[item]
                                             
    def __len__(self):
        return len(self.labels)
```

```python
# 全连接网络
class NN(nn.Module):
    def __init__(self):
        super(NN, self).__init__()
        
        self.model = nn.Sequential(
            nn.Linear(in_features=1024+210+9, out_features=1024),
            nn.ReLU(),
            nn.Linear(in_features=1024, out_features=512),
            nn.ReLU(),
            nn.Linear(in_features=512, out_features=64),
            nn.ReLU(),
            nn.Linear(in_features=64, out_features=32),
            nn.ReLU(),
            nn.Linear(in_features=32, out_features=2),
            nn.Softmax(1)
        )
    def active(self, x):
        for i in range(x.shape[0]):
            mid = ((x[i,:] - x)**2).sum(dim=1, keepdim=True).T
            if i < 1:
                dm = mid
            else:
                dm = torch.cat((dm, mid), dim=0)
        DM = torch.exp(-self.radiu**2 * dm)

    def forward(self, inputs):
        inputs1 = inputs[:,:1024]
        inputs2 = inputs[:,1024:]
        outputs = 0.5*(self.model1(inputs1) + self.model2(inputs2))
        return outputs
```

```python
# RBF块
class new_Linear(nn.Module):
    def __init__(self, in_features, out_features, bias=True):
        super(new_Linear, self).__init__()
        
        self.weight = Parameter(torch.Tensor(in_features, out_features))
        if bias:
            self.bias = Parameter(torch.Tensor(out_features))
        else:
            self.register_parameter('bias', None)

        # 初始化权值
        self.weight.data.uniform_(-1, 1)
        if bias is not None:
            self.bias.data.uniform_(-1, 1)
        

    def forward(self, inputs):
        for i in range(inputs.shape[0]):
            # mid = ((inputs[i,:] - inputs)**2).sum(dim=1, keepdim=True).T
            mid = torch.norm((inputs[i,:] - inputs), p=2, dim=1, keepdim=True).T
            if i < 1:
                dm = mid
            else:
                dm = torch.cat((dm, mid), dim=0)
        DM = torch.exp(-0.01 * dm)
        outputs = DM @ inputs @ self.weight + self.bias
            
        return outputs
        
```

```python
# RBF网络
class RBF_NN(nn.Module):
    def __init__(self):
        super(RBF_NN, self).__init__()
        
        # 全连接层
        self.model = nn.Sequential(
            rbf(in_features=1243, out_features=512),
            nn.ReLU(),
            rbf(in_features=512, out_features=128),
            nn.ReLU(),
            rbf(in_features=128, out_features=16),
            nn.ReLU(),
            rbf(in_features=16, out_features=2),
            nn.Softmax(1)
        )
        
    def forward(self, inputs):
        outputs = self.model(inputs)
        return outputs
```

```python
# 全连接网络
class new_NN(nn.Module):
    def __init__(self):
        super(new_NN, self).__init__()
        
        self.model = nn.Sequential(
            new_Linear(in_features=1024+210+9, out_features=1024),
            nn.ReLU(),
            new_Linear(in_features=1024, out_features=512),
            nn.ReLU(),
            new_Linear(in_features=512, out_features=64),
            nn.ReLU(),
            new_Linear(in_features=64, out_features=32),
            nn.ReLU(),
            new_Linear(in_features=32, out_features=2),
            nn.Softmax(1)
        )

    def forward(self, inputs):
        outputs = self.model(inputs)
        return outputs
```

```python
# 损失函数
class Loss(nn.Module):
    def __init__(self):
        super(Loss, self).__init__()

    def forward(self, inputs, targets):
        loss = -(targets * torch.log(inputs) + (1 - targets) * torch.log(1 - inputs))
        return loss.mean()
```

```python
# 加载数据集、设置网络训练的参数
train_data = Mydata(data_train, data_label)
train_dataloader = DataLoader(train_data, batch_size=32, shuffle=True)
net = new_NN()
loss_fn = torch.nn.CrossEntropyLoss()
if torch.cuda.is_available():
    net = net.cuda()
    loss_fn = loss_fn.cuda()
optimizer = torch.optim.Adam(net.parameters(), lr=0.001)
epoch = 50
```

```python
# 训练模型，并保存每一轮的总损失和精度
train_loss = []
train_acc = []
for i in range(epoch):
    train_loss_epoch = []
    train_acc_epoch = 0
    for data in train_dataloader:
        inputs, labels = data
        if torch.cuda.is_available():            
            inputs = inputs.cuda()
            labels = labels.cuda()
        outputs = net(inputs)
        loss = loss_fn(outputs, labels)
        optimizer.zero_grad()
        loss.backward()
        optimizer.step()
        train_loss_epoch.append(loss.item())
        train_acc_epoch += (outputs.argmax(1)==labels).sum().item()
    train_loss.append(np.sum(train_loss_epoch)/len(train_data))
    train_acc.append(train_acc_epoch/len(train_data)*100)
```

```python
plt.subplot(2, 2, 1)
plt.plot(train_loss[:])
plt.title("train_loss")
plt.subplot(2, 2, 2)
plt.plot(train_acc[:])
plt.title("train_acc")
```
